```python
# ==============================================================
# Classes in Python: objects and blueprints
# ==============================================================
# Python is an object-oriented language.
# Classes define *blueprints* for objects — useful when representing a "thing",
# such as a physical object in computational physics.

import numpy as np

class Person:
    def __init__(self, height, width=40):
        # The constructor (__init__) defines how to create the object.
        # "self" refers to *this* specific instance of the class.
        self.w = width   # attribute: width (cm)
        self.h = height  # attribute: height (cm)

    def volume(self): # "self" is how a class refers to itself. It’s always the first argument in methods, but you don’t pass it manually when calling the method.
        """Approximate a person as a cylinder and compute their volume."""
        return np.pi * (self.w / 2.)**2 * self.h # <- note how "self" makes reference to w and h living inside the object.

# --------------------------------------------------------------
# Create an object: an *instance* of the Person class
christian = Person(180)

# Access attributes directly
print(christian.w)         # 40 (default)
print(christian.volume())  # initial volume

# --------------------------------------------------------------
# Attributes can be modified after creation
christian.w = 50
print(christian.w)
print(christian.volume())

christian.h = 185
print(christian.h)
print(christian.volume())
```

```python
# ==============================================================
# Private members in Python (sort of…)
# ==============================================================
# Python supports *name mangling* for attributes starting with "__".
# This is *not* true privacy like in C++ — it's just a convention.
# You *can* make attributes "private", but I advise against it:
# - It leads to subtle behavior and unnecessary boilerplate.
# - Python relies more on *conventions* than enforcement.

class PrivatePerson:
    def __init__(self, height, width=40):
        self.w = width           # public attribute
        self.__h = height        # "private" attribute via name mangling

    def height(self):            # getter method
        return self.__h

    @property
    def h(self):                 # getter as a property
        return self.__h
    
    @h.setter
    def h(self, height):         # setter for the property
        self.__h = height

    def volume(self):
        """Approximate volume assuming a cylindrical shape."""
        return np.pi * (self.w / 2.)**2 * self.__h

# --------------------------------------------------------------
# Using the class
christian = PrivatePerson(180)

print(christian.height())   # ok: getter method
print(christian.h)          # ok: property getter
# print(christian.__h)      # not ok: attribute name is mangled
print(christian.w)          # ok: public attribute
print(christian.volume())   # ok: method uses private attribute

christian.h = 120           # ok: uses property setter
print(christian.volume())

# --------------------------------------------------------------
# Gotcha: this works but doesn't do what you think!
christian.__h = 100         # creates a *new* public attribute "__h"
print(christian.volume())   # still uses the *original* mangled attribute!
print(christian.__h)        # prints 100 — but it's NOT the same value as h
print(christian.h)          # uses the real private attribute via property


```

```python
# ==============================================================
# Inheritance: classes can build on other classes
# ==============================================================
# A child class automatically inherits attributes and methods from its parent.
# This is useful when you have a "special case" of an existing concept.

class Teacher(Person):  # A Teacher *is-a* Person
    def __init__(self, height, width, students=None):
        # Call the parent (Person) constructor to set height/width
        super().__init__(height, width)

        # Avoid mutable defaults trap: create a new list if needed
        if students is not None and isinstance(students, list):
            self.students = students
        else:
            self.students = []

    # ----------------------------------------------------------
    # Make the class iterable by implementing __iter__
    # This lets us directly loop over Teacher objects.
    def __iter__(self):
        return iter(self.students)

# --------------------------------------------------------------
# Example: a Teacher with a list of students
names = ['Mårten', 'Anna', 'Raoul', 'Nadia']
christian = Teacher(180, 45, names)

print(christian.students)   # inherited + new attribute
print(christian.volume())   # inherited method from Person

# Because we implemented __iter__, we can now loop directly:
for k in christian:
    print(k)
```

```python
# ==============================================================
# Multiple classes inheriting from the same base
# ==============================================================
# Both Teacher and Student inherit from Person.
# This is called *hierarchical inheritance*:
#   Teacher(Person)
#   Student(Person)

class Student(Person):  # A Student *is-a* Person too!
    def __init__(self, height, width, name='Kalle', classes=None, grades=None):
        super().__init__(height, width)
        self.name = name  # immutable string: safe to assign directly

        # Set the student's classes (list of course codes)
        if classes is not None and isinstance(classes, list):
            self.classes = classes
        else:
            self.classes = []

        # Set the student's grades (list of numbers)
        if grades is not None and isinstance(grades, list):
            self.grades = grades
        else:
            self.grades = []

        # Warn if number of classes and grades don't match
        if len(self.grades) != len(self.classes):
            print(f"Warning: student {name} has mismatched classes and grades!")

    # Make Student iterable over grades
    def __iter__(self):
        return iter(self.grades)

# --------------------------------------------------------------
# Create random students: random height (150–200 cm),
# random grades (0 to 10), fixed classes for this demo.
import random
students = [
    Student(
        150 + 50 * random.random(),    # random height
        35,                            # width (cm)
        n,                             # student name
        ['NUMA01', 'FYSN33'],          # enrolled courses
        [round(10 * random.random()),  # random grade for NUMA01
         round(10 * random.random())]  # random grade for FYSN33
    )
    for n in names
]

# --------------------------------------------------------------
# A Teacher holds students, but Teacher.students can hold *any* object.
christian = Teacher(180, 40, students)
```

```python
# ==============================================================
# Iterating over nested objects
# ==============================================================
# Because Teacher implements __iter__, we can loop directly over students.
print("Here is an overview of Christian's students:")

for s in christian:  # s is a Student object
    print(f"{s.name} is {s.h:.2f} cm tall and has taken the courses:")
    for c in s.classes:  # loop over the list of courses for this student
        print(f"  {c}")
```

```python
# ==============================================================
# Using built-in Python tools on custom objects
# ==============================================================
# Because we made Student iterable over grades (__iter__),
# we can directly use built-in functions like mean(), sum(), list(), etc.

from statistics import mean

print("The students have average grades:")

for s in christian:  # s is a Student object
    # mean(s) works because Student.__iter__ returns an iterator over grades
    print(f"Average grade for {s.name} is {mean(s)} ({s.grades})")
```

```python
# ==============================================================
# Bonus: Average grade per course for all of Christian's students
# ==============================================================
from statistics import mean

print("Average grade per course for Christian's students:")

# Collect all distinct courses among the students.
# This is a *set comprehension*, which automatically removes duplicates.
all_courses = {c for s in christian for c in s.classes}

# Loop over all courses in alphabetical order.
for course in sorted(all_courses):
    # Build a list of grades for the current course.
    # This comprehension has:
    #   - The *result expression* (g) at the start.
    #   - Two *for-loops* to iterate students and their courses.
    #   - A *filter* at the end, so only matching courses are kept.
    course_grades = [
        g # <- Result expression
        for s in christian # <- Iterate over Christian's students
        for c, g in zip(s.classes, s.grades) # Iterate over student's classes and grades
        if c == course # <- The filter
    ]

    # Only compute the mean if there are any grades for this course.
    if course_grades:
        print(f"  {course}: {mean(course_grades):.2f} ({course_grades})")
    else:
        print(f"  {course}: no grades available")
```
