```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import diags
from scipy.sparse.linalg import spsolve
# Parameters
dx = 1.0 # Spatial step size
L = 500.    # Length of the domain
Nx = int(L / dx)    # Number of spatial points
dt = 0.1   # Time step
T = 150.0     # Total simulation time
Nt = int(T / dt) # Number of time steps
V0 = 0.7 # Single potential barrier height
a = 250 # Start of V0 in x-coordinates
b = 260 # End of V0 in x-coordinates
x0 = 200 # Start of wave packet
sigma = 20 # Width of wave packet
q = 2 # Wave packet momentum

# Space grid
x = np.linspace(0, L, Nx)

# Initial wavefunction (e.g., Gaussian wave packet)
psi = np.exp(1j*q*x)*np.exp(-(x - x0)**2/(2*sigma**2))

# Dirichlet
psi[0] = 0
psi[-1] = 0

# Potential barrier
V = np.zeros(Nx)
for i in range(Nx):
    if i*dx > a and i*dx < b:
        V[i] = V0

# Construct the A matrix.
coeff = -1j * dt / (2.0 * dx * dx)  # Off-diagonal coefficient
diag_main = 1 + 1j * dt / (dx * dx) + 1j * dt * V  # Including the potential term
diag_upper = np.full(Nx-1 , coeff)  # Off-diagonal upper
diag_lower = np.full(Nx-1, coeff)  # Off-diagonal lower

# Use scipy diags to construct the matrix A
A = diags([diag_lower, diag_main, diag_upper], offsets=[-1, 0, 1], shape=(Nx, Nx)).toarray()

# Take the inverse
Ainv = np.linalg.inv(A)

# Plot at a given n
def plotPsi(ax, x, psi, n):
    ax.plot(x, psi)
    ax.set_title(r'$t = $'+str(n*dt))

# Set up a plotting frame
fig, axs = plt.subplots(2,3)
# Time evolution
for n in range(Nt+1):
    if n == 0:
        plotPsi(axs[0,0],x,abs(psi)**2,n)
    if n == 250:
        plotPsi(axs[0,1],x,abs(psi)**2,n)
    if n == 500:
        plotPsi(axs[0,2],x,abs(psi)**2,n)
    if n == 750:
        plotPsi(axs[1,0],x,abs(psi)**2,n)
    if n == 1000:
        plotPsi(axs[1,1],x,abs(psi)**2,n)
    if n == 1500:
        plotPsi(axs[1,2],x,abs(psi)**2,n)
    
    psi = np.dot(Ainv,psi)
    psi /= np.linalg.norm(psi)
    psi[0] = 0
    psi[-1] = 0

    
# Style the figure a bit and save it
for ax in axs.flat:
    ax.set(xlabel=r'$x$', ylabel=r'$|\Psi(x)|^2$')
    ax.label_outer()
fig.suptitle(r'Schrödinger eqn. Implicit Euler')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import diags

# Parameters
dx = 1.0 # Spatial step size
L = 500.    # Length of the domain
Nx = int(L / dx)    # Number of spatial points
dt = 0.1   # Time step
T = 150.0     # Total simulation time
Nt = int(T / dt) # Number of time steps
V0 = 0.7 # Single potential barrier height
a = 250 # Start of V0 in x-coordinates
b = 260 # End of V0 in x-coordinates
x0 = 200 # Start of wave packet
sigma = 20 # Width of wave packet
q = 2 # Wave packet momentum

# Space grid
x = np.linspace(0, L, Nx)

# Initial wavefunction (e.g., Gaussian wave packet)
psi = np.exp(1j*q*x)*np.exp(-(x - x0)**2/(2*sigma**2))

# Dirichlet
psi[0] = 0
psi[-1] = 0

# Potential barrier
V = np.zeros(Nx)
for i in range(Nx):
    if i*dx > a and i*dx < b:
        V[i] = V0

# Matrices A and B
alpha = 1j * dt / (2 * dx**2)
diagonal_A = 1 + 1j * dt / dx**2 + 1j * dt * V
off_diagonal_A = -0.5j * dt / dx**2

A = diags([off_diagonal_A, diagonal_A, off_diagonal_A], offsets=[-1, 0, 1], shape=(Nx, Nx)).toarray()

diagonal_B = 1 - 1j * dt / dx**2 - 1j * dt * V
off_diagonal_B = 0.5j * dt / dx**2
B = diags([off_diagonal_B, diagonal_B, off_diagonal_B], offsets=[-1, 0, 1], shape=(Nx, Nx)).toarray()

Ainv = np.linalg.inv(A)
M = np.dot(Ainv,B)

# Set up a plotting frame
fig, axs = plt.subplots(2,3)
# Time evolution
for n in range(Nt+1):
    if n == 0:
        plotPsi(axs[0,0],x,abs(psi)**2,n)
    if n == 250:
        plotPsi(axs[0,1],x,abs(psi)**2,n)
    if n == 500:
        plotPsi(axs[0,2],x,abs(psi)**2,n)
    if n == 750:
        plotPsi(axs[1,0],x,abs(psi)**2,n)
    if n == 1000:
        plotPsi(axs[1,1],x,abs(psi)**2,n)
    if n == 1500:
        plotPsi(axs[1,2],x,abs(psi)**2,n)
    
    psi = np.dot(M,psi)
    psi[0] = 0
    psi[-1] = 0


# Style the figure a bit and save it
for ax in axs.flat:
    ax.set(xlabel=r'$x$', ylabel=r'$|\Psi(x)|^2$')
    ax.label_outer()
fig.suptitle(r'Schrödinger eqn. Crank-Nicolson')

```
