```python
# ==============================================================
# Basic data types: integers and floats
# ==============================================================
a = 42
b = 3.14
million = 1_000_000        # underscores improve readability
tiny = 2.5e-6              # 2.5 × 10^-6 (scientific notation)
big  = 7.2e3               # 7.2 × 10^3

print(type(a), a)
print(type(b), b)
print(million, tiny, big)
```

```python
# ==============================================================
# Basic arithmetic operators
# ==============================================================
x = 7
y = 2

print("x + y =", x + y)     # addition
print("x - y =", x - y)     # subtraction
print("x * y =", x * y)     # multiplication
print("x / y =", x / y)     # true division -> always returns float
print("x ** y =", x ** y)   # exponentiation (x to the power y)
print("x % y =", x % y)     # remainder (modulo)
```

```python
# ==============================================================
# Rounding and number formatting
# ==============================================================
x = 2.5
y = 3.5
print("round(2.5) =", round(x))  # ties-to-even ("banker's rounding")
print("round(3.5) =", round(y))

val = 1234.56789
print(f"default:     {val}")       # default formatting, shows full precision
print(f"3 decimals:  {val:.3f}")   # fixed-point notation with 3 decimals
print(f"sci (3 sig): {val:.3e}")   # scientific notation with 3 significant figures
```

```python
# ==============================================================
# Basic data types: booleans (True / False)
# ==============================================================
a = True
if a:  # if-statement: runs only when the condition is True
    print('a is', a)

b = False
if not b:  # 'not' flips the boolean value
    print('b is', b)
```

```python
# ==============================================================
# Boolean logic: and, or, not
# ==============================================================
p, q = True, False

print("p and q ->", p and q)   # True only if both p AND q are True
print("p or q  ->", p or q)    # True if at least one of p OR q is True
print("not p   ->", not p)     # Logical negation: flips True <-> False
```

```python
# ==============================================================
# Basic data types: strings
# ==============================================================
s = 'foo'
print(type(s))    # type() tells us the object type. Note: function calls can be nested.

s2 = "bar"
print(type(s2))

# Concatenate strings (add them together)
print(s + s2)
```

```python
# ==============================================================
# Strings are sequences of characters
# ==============================================================
fysn33 = 'Hello FYSN33!'

# Indexing works like lists: positions start at 0
print(fysn33[0])  # 'H'
print(fysn33[1])  # 'e'
print(fysn33[2])  # 'l'
print(fysn33[3])  # 'l'
print(fysn33[4])  # 'o'
# etc...

# Negative indices count from the end:
print(fysn33[-1])  # '!' (last character)
print(fysn33[-2])  # '3' (second-to-last character)
```

```python
# ==============================================================
# Strings are iterable (you can loop over their characters)
# ==============================================================
for s in fysn33:
    print(s)

# Strings are immutable: you cannot change them in place.
# The following line would cause an error:
#fysn33[0] = 'g'
```

```python
# ==============================================================
# Converting between strings and numbers
# ==============================================================
stringNumber = '10'
number = 10

# Strings and numbers are NOT the same type
if stringNumber == number:
    print('A string number is a number')
else:
    print('A string number is NOT a number')

# Convert string -> integer
if int(stringNumber) == number:
    print('I can convert from string to integer')

# Convert integer -> string
if stringNumber == str(number):
    print('I can convert from integer to string')
```

```python
# ==============================================================
# Useful string methods: example with split()
# ==============================================================
# Python strings have many built-in methods. See:
# https://docs.python.org/3/library/string.html

# Example: split a string into parts based on a delimiter (here: ',')
myString = '1,5,3'
tokens = myString.split(',')   # ['1', '5', '3']
print(tokens)
# Convert the string tokens into floats
x = float(tokens[0])
y = float(tokens[1])
z = float(tokens[2])

print(x + y + z)  # = 9.0
```

```python
# ==============================================================
# Functions and imports: using Python libraries
# ==============================================================
# Python has a huge ecosystem of libraries — use them!
# To find libraries:
#   - Official Python Package Index (PyPI): https://pypi.org
#   - Search the docs: https://docs.python.org/3/library/
#   - Or yes, even ChatGPT can suggest relevant libraries.
#   - See: https://chatgpt.com/share/68ac217c-6ab8-8011-9b3e-fde428d8f2c2

import math

print(math.log(2))                # natural logarithm
print(math.sin(2. * math.pi))     # <- why is this not exactly zero? What can you do?

# Comparing floats directly can be tricky
if math.sin(2. * math.pi) == 0:
    print('sin(2pi) = 0')

# Better: allow for small rounding errors
if abs(math.sin(2. * math.pi)) < 1e-9:
    print('It\'s close enough!')   # <- why did I use double quotes here?

# Hint: math also has an 'isclose' method — find it and try it.

closeenough = math.isclose(math.sin(2 * math.pi), 0, abs_tol=1e-9)
if closeenough:
    print("Yeah it's close!")
# Alternative import styles:
# from math import cos
# print(cos(2. * math.pi))

# You can always ask for help on modules and functions:
#help(math)
```

```python
# ==============================================================
# Defining your own functions (reuse code!)
# ==============================================================
def foo(arg):
    print('I love ' + str(arg) + '!')

foo('FYSN33')   # Good

foo(42)  # Not so good — this would fail, since 42 is not a string.
```

```python
# ==============================================================
# Checking / sanitizing your input
# ==============================================================
def bar(arg):
    if isinstance(arg, str):  # check if arg is a string
        print('I love ' + arg + '!')
    else:
        print('Please input only strings to bar()')

bar('FYSN33')   # OK
bar(42)        # Not OK — triggers the fallback branch
```

```python
# ==============================================================
# Why *not* silently convert types
# ==============================================================
# It's tempting to automatically convert arguments to the "right" type,
# but this can hide mistakes and produce unexpected results.
# It's better if a function is precise about what it expects.

# Example: return the "sum" of all elements of a list (strings only!)
def sumList(arg):
    ret = str(arg[0])
    for a in arg[1:]:   # <- lists can be sliced
        ret += str(a)   # string concatenation
    return ret

print(sumList(['I', 'Love', 'FYSN33', '!']))  # "ILoveFYSN33!"

# Uncomment to see surprising behavior:
print(sumList([1, 2, 3, 4, 5]))  # "12345" — probably not what you intended!
```

```python
# ==============================================================
# Default arguments and more complex logic
# ==============================================================
# Task: Write a function that takes one required argument (arg1)
# and one optional argument (arg2).
# If both are strings, print: "I love <arg1> and <arg2>!"
# Otherwise, just print: "I love <arg1>!"
# There are many ways to implement this — try finding another later!

def baz(arg1, arg2=None):
    ret = ''

    # Check that arg1 is a string
    if isinstance(arg1, str):
        ret += 'I love ' + arg1
    else:
        print('Please input only strings to baz()')
        return ret   # return early — no need to check arg2

    # Add arg2 if provided AND it's a string
    if arg2 is not None and isinstance(arg2, str):  # multiple conditions, evaluated left to right
        ret += ' and ' + arg2

    return ret + '!'

# Examples
print(baz('FYSN33'))             # "I love FYSN33!"
print(baz('FYSN33', 'ice cream'))# "I love FYSN33 and ice cream!"
print(baz(42))                   # Invalid first argument
print(baz('FYSN33', 42))         # Arg2 ignored since it's not a string
```

```python
# ==============================================================
# Bonus: A tighter, more Pythonic implementation
# ==============================================================
# You can explore this implementation during the workshop.
# - Are there features you don't recognize? (e.g. type hints, f-strings)
# - How do they improve readability and safety?
# - How does the behavior differ from the implementation above?

def baz(arg1: str, arg2: str | None = None) -> str:
    """
    Return a string expressing love for one or two things.
    Raises a TypeError if arg1 is not a string.
    """

    if not isinstance(arg1, str):
        raise TypeError("baz() requires arg1 to be a string")

    # Use f-strings for clarity and readability
    return f"I love {arg1} and {arg2}!" if isinstance(arg2, str) else f"I love {arg1}!"

# ==============================================================
# Extra bonus: This is also a valid implementation.
# Figure out how it works — and discuss:
# - Is this actually a good idea?
# - When might "clever" code become harmful?
# ==============================================================
# baz = lambda arg1, arg2=None: (
#     f"I love {arg1}" + (f" and {arg2}" if isinstance(arg2, str) else "") + "!"
#     if isinstance(arg1, str) else
#     (_ for _ in ()).throw(TypeError("baz() requires arg1 to be a string"))
# )
```

```python
# ==============================================================
# while-loops, break, and continue
# ==============================================================
# A while-loop repeats until a condition is met
i = 10
while i > 0:
    print(i)
    i -= 1

# --------------------------------------------------------------
# break: exit the loop early, regardless of the condition
print('......')
i = 10
while True:  # infinite loop unless we break
    print(i)
    i -= 1
    if i < 1:
        break  # stop looping when i < 1

# --------------------------------------------------------------
# continue: skip the rest of the current iteration
print('......')
i = 10
while i > 0:
    if i == 5:
        i -= 1  # <- what would happen if we skipped this line?
        continue  # jump directly to the next iteration
    print(i)
    i -= 1

# --------------------------------------------------------------
# continue also works in for-loops
print('......')
for i in range(10):
    if i + 1 == 5:
        continue  # skip printing 5
    print(i + 1)   # why +1? Because range(10) gives 0..9, so we shift to 1..10. Neater is: range(1,11)

# ==============================================================
# Challenge question for this cell
# ==============================================================
# Write a loop that skips multiples of 3 but stops at the first number divisible by 7
```

```python
# ==============================================================
# Generating lists in different ways
# ==============================================================
# Different ways to build the same or similar lists — from explicit to compact.

# 1. Enter values manually (clear but not scalable)
a = [0, 1, 2, 3, 4, 5, 6, 7]

# 2. Use a for-loop and append
b = []
for i in range(8):
    b.append(2 * i)

# 3. Use a list comprehension: a compact, Pythonic way to build lists
c = [3 * i for i in range(8)]

# 4. List comprehension with a filter: skip i == 3
d = [4 * i for i in range(8) if i != 3]

# 5. Generate a sequence directly using range()
e = list(range(8))  # simplest way to make [0, 1, 2, 3, 4, 5, 6, 7]

print(a, b, c, d, e)
```

```python
# ==============================================================
# Looping over multiple lists and getting indices
# ==============================================================
# zip() lets you loop over multiple lists in parallel:
for an, bn in zip(a, b):
    print(an, bn)  # pairs elements from a and b: (a[0], b[0]), (a[1], b[1]), ...

print('........')

# enumerate() gives you both the index and the value while looping:
for i, an in enumerate(a):
    print(i, an)  # prints index i and value a[i]
```

```python
# ==============================================================
# Dictionaries: fast lookup tables (key → value)
# ==============================================================
# Dicts map "keys" to "values". Think of them as smart lookup tables.
d = {'a': 1, 'b': 12, 'c': 26}
print(type(d), d)

# Access a value by its key:
print(d['b'])   # 12

# Add a new key/value pair:
d['z'] = 42
print(d['z'])   # 42

# --------------------------------------------------------------
# Deleting entries safely
# Always check if a key exists before deleting it.
if 'z' in d:
    del d['z']

# --------------------------------------------------------------
# Handling missing keys using try/except:
try:  # <- This is a try/except block
    print(d['z'])   # Key no longer exists
except KeyError:  # <- Catches only missing-key errors
    print('z is not a valid key!')

# For a full list of built-in exception types, see:
# https://docs.python.org/3/library/exceptions.html
```

```python
# ==============================================================
# Looping over dictionaries
# ==============================================================
# By default, looping over a dict gives you its keys:
for k in d:
    print(k, d[k])   # key and corresponding value

# The more Pythonic way: use .items() to get keys and values directly
for k, v in d.items():
    print(k, v)

# You can also get the keys explicitly (less common):
print(d.keys())

# Or loop directly over the values (also less common):
for v in d.values():
    print(v)
```

```python
# ==============================================================
# Common traps: infinities and NaN (Not a Number)
# ==============================================================
import math

# Create positive/negative infinity and NaN
pos_inf = float('inf')
neg_inf = float('-inf')
nan = float('nan')

print("isfinite:", math.isfinite(1.0), math.isfinite(pos_inf), math.isfinite(nan))
print("isnan:", math.isnan(nan))

# --------------------------------------------------------------
# Beware: any comparison involving NaN is False, even with itself!
print("nan == nan ?", nan == nan)        # False
print("Use math.isnan(nan) instead:", math.isnan(nan))
```

```python
# ==============================================================
# Trap: Floating-point precision
# ==============================================================
# Some decimal numbers cannot be represented exactly in binary.
# This leads to small rounding errors when working with floats.
import math
x = 0.1 + 0.2
print("0.1 + 0.2 =", x)           # Not exactly 0.3!
print("Direct comparison:", x == 0.3)      # False
print("Safe comparison:", math.isclose(x, 0.3))  # True
```

```python
# ==============================================================
# Another common trap: mutable default arguments
# ==============================================================
# Problem: default arguments are evaluated only ONCE, when the function is defined.
# If the default is mutable (like a list), it persists across calls!

# BUG: bucket persists across calls
def append_bad(x, bucket=[]):
    bucket.append(x)
    return bucket

b1 = append_bad("a")
print(b1)   # ['a']
b2 = append_bad("b")
print(b2)   # ['a', 'b']  <-- surprising if you expected ['b']!

# We can inspect the default argument of our method and see that
# it has persisted:
print('Default argument: ',append_bad.__defaults__)

# --------------------------------------------------------------
# Correct pattern: use a sentinel (None) and create a new list per call
def append_good(x, bucket=None):
    if bucket is None:
        bucket = []      # create a new list each time
    bucket.append(x)
    return bucket

g1 = append_good("a")
print(g1)   # ['a']
g2 = append_good("b")
print(g2)   # ['b']  <-- now behaves as expected
```

```python
# ==============================================================
# Trap: copying vs referencing
# ==============================================================
a = [1, 2, 3]
b = a          # b is just another name for the same list!
b[0] = 99
print(a)       # [99, 2, 3] <-- maybe unexpected

# Correct way to make a shallow copy:
b = a.copy()
b[0] = 42
print(a)       # [99, 2, 3] unchanged
print(b)       # [42, 2, 3]
```

```python
# ==============================================================
# Trap: chained comparisons
# ==============================================================
x = 5
print(1 < x < 10)     # True  <-- checks both 1 < x AND x < 10
print(1 < x > 10)     # False <-- because x > 10 fails

# Equivalent to:
print(1 < x and x < 10)

# Recommendation: Stick to the verbose version, it is clearer.
```

```python
# ==============================================================
# Common trap/feature: "truthy" and "falsy" values in Python
# ==============================================================
# In conditionals, Python automatically converts objects to True or False.
# Some values are treated as "falsy" (behave like False) and others as "truthy" (behave like True).

samples = [
    0, 1, -1, 0.0, 0.0001, 1e-16, 1e-400,
    "", "0", "True", "False",
    [], [0], (), (1,), {}, {"a": 1}, set(), {1},
    None
]

# Inspect the truthyness:
for x in samples:
    print(repr(x), "-> bool =", bool(x))
```

```python
# ==============================================================
# Common pattern: short-circuiting in boolean expressions
# ==============================================================
# In Python, boolean operators (and/or) are evaluated left to right.
# The right-hand side (RHS) is only evaluated if needed:
# - For "and", if the left side is False, RHS is skipped.
# - For "or",  if the left side is True,  RHS is skipped.

def side_effect(tag):
    print("side_effect({}) ran".format(tag))
    return True

print("\nShort-circuiting:")

# RHS NOT called because LHS is False
print("False and side_effect ->", False and side_effect("AND"))

# RHS NOT called because LHS is True
print("True or side_effect ->", True or side_effect("OR"))

# RHS IS called because LHS is True and "and" needs both sides
print("True and side_effect ->", True and side_effect("AND"))

# RHS IS called because LHS is False and "or" needs RHS to decide
print("False or side_effect ->", False or side_effect("OR"))
```

```python
# ==============================================================
# Basic file I/O: reading and writing text files
# ==============================================================
# Use open() to access files. Second argument sets the mode:
#   'r' = read   |   'w' = write (overwrites!)   |   'a' = append
# Full docs: https://docs.python.org/3/library/functions.html#open

# Avoid doing this:
f = open('myfile.txt', 'w')
f.write('some text')
f.close()
# If you forget to close(), changes might not be saved!

# --------------------------------------------------------------
# Recommended: use a context manager — closes automatically
with open('myfile.txt', 'w') as f:
    f.write('1,2,3,4\n5,6,7,8') # <- \n is a newline character
# !!BEWARE!! Using 'w' will overwrite an existing file without
# warning!
# Inspect the file in your favourite reader.
# --------------------------------------------------------------
# Reading files line by line:
with open('myfile.txt', 'r') as f:
    for l in f:
        data = l.strip().split(',')  # strip() removes trailing \n
        a, b, c, d = map(float, data)  # convert all at once
        print(a + b + c + d)
```
